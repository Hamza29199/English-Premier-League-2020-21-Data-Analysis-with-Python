## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

## Load Data

```python
epl_df = pd.read_csv('EPL_20_21.csv')
```

```python
epl_df.head(10)
```

```python
epl_df.describe()
```

```python
epl_df.info()
```

```python
#total num of null values?
epl_df.isna().sum()
```

```python
##two new cols added

epl_df['Mins Per Match'] = (epl_df['Mins']/epl_df['Matches']).astype(int)
epl_df['Goals Per Match'] = (epl_df['Goals']/epl_df['Matches']).astype(float)
```

```python
epl_df.head()
```

```python
total_season_goals = epl_df['Goals'].sum(axis=0)
total_season_goals
```

```python
total_season_pen_goals = epl_df['Penalty_Goals'].sum(axis=0)
total_season_pen_goals
```

```python
total_season_pen_atts = epl_df['Penalty_Attempted'].sum(axis=0)
total_season_pen_atts
```

```python
#Pie chart viz for penalties missed vs scored

penalties_no_scored = total_season_pen_atts - total_season_pen_goals
data = [penalties_no_scored, total_season_pen_goals]
labels = ['Missed Pens', 'Scored Pens']

color = sns.color_palette('Set3')
plt.pie(data, labels=labels, colors=color, autopct="%.02f")
plt.show()
```

```python
#Most players from which countries?

most_common_nationality = epl_df.groupby('Nationality').size().sort_values(ascending=False)
most_common_nationality.head(10).plot(kind='bar', color=sns.color_palette('magma'))
```

```python
#Most represented clubs by num of players

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('magma'))
```

```python
epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette('magma'))
```

```python
under_20 = epl_df[epl_df['Age'] < 20]
from_20_to_25 = epl_df[(epl_df['Age'] >=20) & (epl_df['Age'] < 25)]
from_25_to_30 = epl_df[(epl_df['Age'] >=25) & (epl_df['Age'] < 30)]
over_30 = epl_df[epl_df['Age'] >=30]
```

```python
data = [len(under_20),len(from_20_to_25),len(from_25_to_30),len(over_30)]
labels = ['<20', '20-25', '25-30', '>=30']
plt.pie(data, labels=labels, colors=sns.color_palette('viridis'), autopct='%.0f')
plt.show()
```

```python
#boxplot to see average age for each club

sns.boxplot(x='Club', y='Age', data=epl_df)
plt.xticks(rotation=90) #so that we see the club names
```

```python
#average age for clubs

num_players = epl_df.groupby('Club').size()
age_avg = (epl_df.groupby('Club')['Age'].sum())/num_players
age_avg.sort_values()
```

```python
#Top 10 assisters

epl_df[['Name', 'Assists']].nlargest(10, columns='Assists')
```

```python
#Top 10 scorers

epl_df[['Name', 'Goals']].sort_values('Goals', ascending=False).head(10)
```

```python
#Plot of Clubs vs Total Assists

assists_by_club = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Assists'].sum())

ax = sns.barplot(x='Club', y='Assists', data = assists_by_club.sort_values('Assists'))
ax.set_xlabel('Clubs', fontsize=20)
ax.set_ylabel('Assists', fontsize=20)

plt.xticks(rotation = 90)
plt.title('Plot of Clubs vs Total Assists', fontsize =30)
plt.show()
```

```python
#Plot of Clubs vs Total Goals

goals_by_club = pd.DataFrame(epl_df.groupby('Club', as_index=False)['Goals'].sum())

ax = sns.barplot(x='Club', y='Goals', data = goals_by_club.sort_values('Goals'))
ax.set_xlabel('Clubs', fontsize=20)
ax.set_ylabel('Goals', fontsize=20)

plt.xticks(rotation = 90)
plt.title('Plot of Clubs vs Total Goals', fontsize =30)
plt.show()
```

```python
#top 15 most yellow-carded players

epl_most_yellowed = pd.DataFrame(epl_df[['Name', 'Yellow_Cards']])

ax = sns.barplot(x='Name', y='Yellow_Cards', data = epl_most_yellowed.sort_values('Yellow_Cards',ascending=False).head(15), palette='rocket')
ax.set_xlabel('Player Name', fontsize=20)
ax.set_ylabel('Yellow Cards', fontsize=20)
plt.xticks(rotation =90)
plt.title('Top 15 Most Yellow Carded Players')
plt.show()
```
